# Cleaning Data

## Cleaning product_df


* product_df

```python
product_df[product_df.product_category_name.isna()]
```

```python
product_df['product_category_name'].fillna(value="No Category", inplace=True)

num_product = ["product_name_lenght", "product_description_lenght", "product_photos_qty", "product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"]

for i in num_product:
  product_df[i].fillna(value=0, inplace=True)
```

```python
product_df.isna().sum()
```

Memastikan kembali bahwa sudah tidak ada missing value.

```python
product_df[product_df["product_category_name"] == "No Category"]
```

## Cleaning order_df

Pada order_df, terdapat banyak missing value pada data. Namun, ketika ditelaah kembali dapat dilihat bahwa missing value pada column tersebut disebabkan karena order yang sedang diproses, dibatalkan, belum dibayar, ataupun order yang belum sampai di customer, sehingga belum ada tanggal pengirimannya.

```python
order_df[order_df.order_approved_at.isna()]
```

```python
order_df[order_df.order_delivered_carrier_date.isna()]
```

```python
order_df[order_df.order_delivered_customer_date.isna()]
```

```python
convert_datatype = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]

for i in convert_datatype:
  order_df[i] = pd.to_datetime(order_df[i])
```

## Eksplorasi product_df dan product_translation_df

```python
product_df.describe(include="all")
```

Untuk data dari product kita dapat mengganti nama dari product category dengan translasi yang sudah disediakan.

```python
product_translation_df.describe(include="all")
```

```python
product_translated_df = pd.merge(
    left = product_df,
    right = product_translation_df,
    how = "left",
    left_on = "product_category_name",
    right_on = "product_category_name"
)

product_translated_df.drop(columns=["product_category_name"], inplace=True)
product_translated_df.rename(columns={
    "product_category_name_english": "product_category_name"
    }, inplace=True)

product_translated_df.head()
```

## Eksplorasi order_df

Untuk data order_df, kita dapat mengurangi waktu dari order_delivered_customer_date dengan order_delivered_carrier_date untuk menghasilkan lama waktunya pengiriman, kemudian dapat kita bandingkan apakah sesuai dengan estimasi waktu pengiriman atau tidak.

```python
delivery_time = order_df["order_delivered_customer_date"] - order_df["order_delivered_carrier_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
order_df["delivery_time"] = round(delivery_time/86400)

order_df.describe(include="all")
```

## Eksplorasi order_df dan customer_df

```python
customer_order_set = set(order_df.customer_id.values)
customer_df["status"] = customer_df["customer_id"].apply(lambda x: "Active" if x in customer_order_set else "Non-Active")
customer_df.sample(5)
```

```python
customer_df.groupby("status").customer_id.nunique()
```

```python
order_cust_df = pd.merge(
    left = order_df,
    right = customer_df,
    how = "left",
    left_on = "customer_id",
    right_on = "customer_id"
)

order_cust_df.sample(5)
```



*   customer yang banyak melakukan reorder


```python
order_cust_df.groupby("customer_unique_id").agg({
    "order_id": "nunique",
    "status": "unique"
}).sort_values(by="order_id", ascending=False)
```

## Eksplorasi order_payment_df dan order_review_df

```python
payrev_df = pd.merge(
    left = order_payment_df,
    right = order_review_df,
    how = "left",
    left_on = "order_id",
    right_on = "order_id"
)

payrev_df.sample(5)
```

```python
payrev_df.groupby("review_score").agg({
    "order_id": "nunique",
    "payment_value": ["min", "max", "sum"]
})
```

```python
cust_payrev_df = pd.merge(
    left = order_cust_df,
    right = payrev_df,
    how = "left",
    left_on = "order_id",
    right_on = "order_id"
)

cust_payrev_df.sample(5)
```

## Eksplorasi order_item_df dan product_translated_df

```python
order_product_df = pd.merge(
    left = order_item_df,
    right = product_translated_df,
    how = "left",
    left_on = "product_id",
    right_on = "product_id"
)

order_product_df.sample(5)
```

```python
order_product_df.groupby("product_category_name").agg({
    "order_id": "nunique",
    "price": "sum"
}).sort_values(by="order_id", ascending=False)
```

## Eksplorasi order_product_df dan seller_df

```python
product_seller_df = pd.merge(
    left = order_product_df,
    right = seller_df,
    how = "left",
    left_on = "seller_id",
    right_on = "seller_id"
)

product_seller_df.sample(5)
```

```python
product_seller_df.groupby("seller_city").agg({
    "order_id": "nunique",
    "price": "sum"
}).sort_values(by="order_id", ascending=False)
```

## Explore all data

```python
all_df = pd.merge(
    left = cust_payrev_df,
    right = product_seller_df,
    how = "left",
    left_on = "order_id",
    right_on = "order_id"
)

all_df.head()
```

```python
all_df.info()
```

```python
all_df.groupby(by=["customer_state","product_category_name"]).agg({
    "order_id": "nunique",
    "price": "sum"
}).sort_values(by="order_id", ascending=False)
```

```python
all_df.groupby(by=["customer_city", "product_category_name"]).agg({
    "order_id": "nunique",
    "price": "sum"
}).sort_values(by="order_id", ascending=False)
```

# Data Visualization

## Apa tipe pembayaran yang banyak digunakan oleh customer?

```python
most_payment_type = all_df.groupby(by="payment_type").order_id.nunique().sort_values(ascending=False).reset_index()
most_payment_type.rename(columns={
    "order_id": "counts"
 }, inplace=True)

most_payment_type.head()
```

```python
fig, ax = figsize=(8, 8)
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(
    x="counts",
    y="payment_type",
    hue="payment_type",
    legend=False,
    data=most_payment_type,
    palette=colors
)

plt.title("Most Used Payment Type")
plt.xlabel(None)
plt.ylabel(None)
plt.show()
```

## Bagaimana tingkat kepuasan customer terhadap platform e=commerce?

```python
cust_satisfaction = all_df.groupby(by="review_score").order_id.nunique().reset_index()
cust_satisfaction.rename(columns={
    "order_id": "counts"
 }, inplace=True)

cust_satisfaction.head()
```

```python
plt.figure(figsize=(10,5))
colors = ["#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#72BCD4"]

sns.barplot(
    y="counts",
    x="review_score",
    hue="review_score",
    legend=False,
    data = cust_satisfaction.sort_values(by="counts", ascending=False),
    palette=colors
)

plt.title("Customer Satisfaction Rating Overall", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

## Bagaimana kinerja produk pada e-commerce?

```python
product_performance = all_df.groupby(by="product_category_name").agg({
    "order_id": "nunique",
    "price": "sum",
    "review_score": "mean"
})

product_performance.rename(columns={
    "order_id": "qty_sold",
    "price": "revenue",
    "review_score": "average_score"
 }, inplace=True)

product_performance.head()
```

```python
fig, ax = plt.subplots(ncols=3, nrows= 1, figsize=(25, 5))
colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="qty_sold", x="product_category_name", data=product_performance.sort_values(by="qty_sold", ascending=False).head(5), ax=ax[0])
ax[0].set_xlabel(None)
ax[0].set_ylabel(None)
ax[0].set_title("By Quantity Sold")
ax[0].tick_params(axis='x', labelsize=8)

sns.barplot(y="revenue", x="product_category_name", data=product_performance.sort_values(by="revenue", ascending=False).head(5), ax=ax[1])
ax[1].set_xlabel(None)
ax[1].set_ylabel(None)
ax[1].set_title("By Revenue")
ax[1].tick_params(axis='x', labelsize=8)

sns.barplot(y="average_score", x="product_category_name", data=product_performance.sort_values(by="average_score", ascending=False).head(5), ax=ax[2])
ax[2].set_xlabel(None)
ax[2].set_ylabel(None)
ax[2].set_title("By Rating")
ax[2].tick_params(axis='x', labelsize=8)

plt.suptitle("Best Products Performance", fontsize=20)
plt.show()
```

## Bagaimana demografi dari customer saat ini?

```python
bystate_df = all_df.groupby(by="customer_state").customer_id.nunique().reset_index()
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)

plt.figure(figsize=(10, 5))
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
sns.barplot(
    y="customer_state",
    x="customer_count",
    data=bystate_df.sort_values(by="customer_count", ascending=False).head(8),
    palette=colors
)

plt.title("Customer Demographics by State")
plt.xlabel("State")
```

* by city

```python
bycity_df = all_df.groupby(by="customer_city").customer_id.nunique().reset_index()
bycity_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)

plt.figure(figsize=(10, 5))
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
sns.barplot(
    y="customer_city",
    x="customer_count",
    data=bycity_df.sort_values(by="customer_count", ascending=False).head(8),
    palette=colors
)

plt.title("Customer Demographics by City")
plt.xlabel(None)
plt.ylabel(None)
```
